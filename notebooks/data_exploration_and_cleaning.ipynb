# MovieLens - Ingestion Batch et Vérification de la Qualité


# ETL Pipeline - MovieLens Dataset avec PySpark


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, isnan, when, count, explode, split, avg, desc
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Initialisation de Spark


```python
spark = SparkSession.builder \
    .appName("BatchETL") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .config("spark.sql.shuffle.partitions", "8") \
    .getOrCreate()
    
spark.sparkContext.setLogLevel("ERROR")
```

## 2. Chargement des fichiers CSV depuis HDFS


```python
ratings_path = "hdfs://namenode:9000/user/movielens/raw/rating.csv"
movies_path = "hdfs://namenode:9000/user/movielens/raw/movie.csv"

ratings_df = spark.read.option("header", True).option("inferSchema", True).csv(ratings_path)
movies_df = spark.read.option("header", True).option("inferSchema", True).csv(movies_path)
```

```python
# Affichage des schémas
print("Schéma des notes:")
ratings_df.printSchema()

print("\nSchéma des films:")
movies_df.printSchema()
```

## 3. Vérification de la qualité des données


```python

def count_nulls(df):
    exprs = []
    for c in df.columns:
        dtype = dict(df.dtypes)[c]
        if dtype in ["float", "double"]:
            exprs.append(count(when(col(c).isNull() | isnan(col(c)), c)).alias(c))
        else:
            exprs.append(count(when(col(c).isNull(), c)).alias(c))
    return df.select(exprs)
```

```python
# Suppression des valeurs nulles
ratings_df = ratings_df.dropna()
movies_df = movies_df.dropna()
```

```python
# Suppression des doublons
ratings_df = ratings_df.dropDuplicates()
movies_df = movies_df.dropDuplicates()
```

```python
# Vérification des valeurs manquantes
print("Valeurs manquantes dans les notes:")
count_nulls(ratings_df).show()

print("\nValeurs manquantes dans les films:")
count_nulls(movies_df).show()
```

## 4. Visualisation de la distribution des notes

```python
# Conversion en pandas pour la visualisation
ratings_pandas = ratings_df.select("rating").limit(50000).toPandas()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(
    ratings_pandas["rating"], 
    bins=10, 
    kde=True, 
    kde_kws={"bw_adjust": 1.5},  # Plus la valeur est grande, plus la courbe est lisse
    color="#4C72B0", 
    edgecolor='black', 
    alpha=0.7
)

plt.title('Distribution des notes', fontsize=16, fontweight='bold')
plt.xlabel('Note', fontsize=12)
plt.ylabel('Fréquence', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(ticks=[0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])
plt.tight_layout()
plt.show()
```

```python
genre_counts = (
        movies_df
        .withColumn("genre", explode(split("genres", r"\|")))
        .groupBy("genre")
        .count()
        .orderBy("count", ascending=False)
    )

# Création d'un graphique en barres pour visualiser le nombre de films par genre
genre_pd = genre_counts.toPandas()
plt.figure(figsize=(12, 6))
sns.barplot(x="genre", y="count", data=genre_pd)
plt.xticks(rotation=45)
plt.title("Nombre de films par genre")
plt.xlabel("Genre")
plt.ylabel("Nombre de films")
plt.tight_layout()
plt.show()
```

```python
# Drop les genres a (no genres listed)
movies_df = movies_df[(movies_df.genres != '(no genres listed)')]
```

```python
# Jointure sur movieId pour récupérer les titres
votes_df = ratings_df.groupBy("movieId").count()
top_movies_df = votes_df.join(movies_df, on="movieId").orderBy(desc("count")).limit(10)

# Conversion en pandas pour affichage
top_movies_pd = top_movies_df.select("title", "count").toPandas()
top_movies_pd["Classement"] = [f"TOP {i+1}" for i in range(len(top_movies_pd))]
top_movies_pd["Titre_affiche"] = top_movies_pd["Classement"] + " - " + top_movies_pd["title"]

# Affichage du graphique
plt.figure(figsize=(10, 6))
sns.barplot(y="Titre_affiche", x="count", data=top_movies_pd, palette="magma")
plt.xlabel("Nombre de votes")
plt.ylabel("Titre du film")
plt.title("Nombre de votes pour les films du Top 10")
plt.tight_layout()
plt.show()
```

```python
# Filtrage des notes extrêmes
ratings_df = ratings_df.filter((col("rating") >= 0.5) & (col("rating") <= 5.0))

# Jointure pour enrichir les données
ratings_enriched_df = ratings_df.join(movies_df, on="movieId", how="inner")

# Partitionnement par userId et sauvegarde dans HDFS (parquet)
ratings_enriched_df.repartition(8) \
    .write \
    .mode("overwrite") \
    .partitionBy("userId") \
    .parquet("hdfs://namenode:9000/user/movielens/clean/ratings_by_user")

print("\n✅ Pipeline ETL terminé. Données sauvegardées dans /user/movielens/clean/ratings_by_user")
```
